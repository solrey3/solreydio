```python
csv_file = "150ishFavoriteAlbums.csv"
```

```python
albums150 = pd.read_csv(csv_file)
albums150
```

```python
favoritealbums_df = pd.DataFrame({"artist": albums150.artist, "album": albums150.album, "wikipedia_url": albums150.wikipedia_url, "release_date": albums150.release_date})
favoritealbums_df.to_json('favoritealbums.json', orient='records')
favoritealbums_df
```
