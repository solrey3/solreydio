```python
csv_file = "150ishFavoriteAlbums.csv"
```

```python
albums150 = pd.read_csv(csv_file)
albums150
```

```python
favoritealbums_df = pd.DataFrame({"album": albums150.album, "wikipedia_url": albums150.wikipedia_url, "artist": albums150.artist, "release_date": albums150.release_date})
favoritealbums_df.to_json('favoritealbums.json', orient='records')
favoritealbums_df
```

```python
#pareto = pd.DataFrame({"rank": albums150.rank, "album": albums150.album, "wikipedia_url": albums150.wikipedia_url, "artist": albums150.artist, "release_date": albums150.release_date})
ranked = albums150.dropna()

ranked = ranked[['rank', 'album', 'wikipedia_url', 'artist', 'release_date']]
# pareto.rank = int(pareto.rank)

ranked = ranked.sort_values(by=['rank'])

ranked = ranked.astype({'rank': 'int32'})

ranked.to_json('ranked.json', orient='records')

ranked
```

```python
artist_album_counts = favoritealbums_df["artist"].value_counts()
artist_group = pd.DataFrame({"Count": artist_album_counts})
artist_group = artist_group.loc[artist_group.Count > 1, :]
artist_group.reset_index(level=0, inplace=True)
artist_group = artist_group.rename(columns={'index': 'artist', 'Count':'count'})
artist_group = artist_group.sort_values(by=['count', 'artist'], ascending=[False, True])
artist_group.to_json('artistcounts.json', orient='records')
artist_group
```
